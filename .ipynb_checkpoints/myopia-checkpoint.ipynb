```python
# Initial imports
import pandas as pd
from sklearn.cluster import KMeans
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading data
file_path = Path("./Resources/myopia.csv")
myopia_df = pd.read_csv(file_path)
myopia_df.head()
```

```python
myopia_df.dtypes
```

```python
# Remove the Myopic Column
myopia_df = myopia_df.drop(columns=["MYOPIC"])
myopia_df.head()
```

```python
# Find null values
for column in myopia_df.columns:
    print(f"Column {column} has {myopia_df[column].isnull().sum()} null values")

myopia_df.head()
```

```python
# Find duplicate entries
print(f"Duplicate entries: {myopia_df.duplicated().sum()}")
```

```python
#SCALING THE DATA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(myopia_df[['AGE', 'SPHEQ', 'AL', 'ACD', 'LT', 'VCD', 'SPORTHR', 'READHR', 'COMPHR', 'STUDYHR', 'TVHR', 'DIOPTERHR', 'MOMMY', 'DADMY']])
print(scaled_data[0:5])
```

```python
components = 3
```

```python
# Initialize PCA model
pca = PCA(n_components= components)

# Get two principal components for the iris data.
myopia_pca = pca.fit_transform(scaled_data)

myopia_pca 
```

```python
# Run t-SNE
tsne = TSNE(learning_rate=250)
tsne_features = tsne.fit_transform(myopia_pca)
```

```python
# Check that dimensions have been reduced to two
tsne_features.shape
```

```python
# Plot the results
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
pca.explained_variance_ratio_.sum()
```

```python
col_names = [f"PC {i}" for i in range(1, len(myopia_pca[0])+1)]
col_names
```

```python
# Transform PCA data to a DataFrame
myopia_pca_df = pd.DataFrame(
    data=myopia_pca, columns= col_names
)
myopia_pca_df.head()
```

```python
# Fetch the explained variance 
pca.explained_variance_ratio_
```

```python
#K Means

inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(myopia_pca_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for Myopia data')
plt.show()
```

```python
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Train the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
clusters = get_clusters(3, myopia_pca_df)

clusters
```

```python
myopia_class = clusters[['class']]
myopia_class
```

```python
# concat dataframes
myopia_class_df = pd.concat([myopia_df, myopia_class], axis=1, ignore_index=True)

myopia_class_df
```
